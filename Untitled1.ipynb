Importing of the necessary library

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn import metrics
```

```python
car_data = pd.read_csv('/content/car_data[1].csv')
```

```python
car_data.head()
```

Observing the shape of the dataset

```python
car_data.shape
```

Collecting the information of all the variables

```python
car_data.info()
```

Checking whether any missing values is there in the dataset

```python
car_data.isnull().sum()
```

```python
car_data.describe()
```

```python
car_data.head(1)
```

```python
import datetime
date_time = datetime.datetime.now()
car_data['Age']= date_time.year - car_data['Year']
car_data.head()
```

Dropping the Year and Car_Name column of the dataset

```python
car_data.drop(['Car_Name','Year'],axis=1,inplace=True)
```

```python
car_data.head()
```

```python
sns.boxplot(car_data['Selling_Price'])
```

```python
sorted(car_data['Selling_Price'],reverse = True)
```

Encoding the value of Fuel_Type, Transmission and Seller_Type as numerical values like 0, 1, 2 and so on

```python
car_data['Fuel_Type'].unique()
```

```python
car_data['Fuel_Type']= car_data['Fuel_Type'].map({'Petrol':0,'Diesel':1,'CNG':2})
car_data['Fuel_Type'].unique()
```

```python
car_data['Seller_Type'].unique()
```

```python
car_data['Seller_Type']= car_data['Seller_Type'].map({'Dealer':0,'Individual':1})
car_data['Seller_Type'].unique()
```

```python
car_data['Transmission'].unique()
```

```python
car_data['Transmission']= car_data['Transmission'].map({'Manual':0,'Automatic':1})
car_data['Transmission'].unique()
```

Store Feature Matrix (Independent variable) in X and Response(target) in Vector Y

```python
x = car_data.drop(['Selling_Price'],axis=1)
y = car_data['Selling_Price']
```

```python
x
```

```python
y
```

Splitting the Dataset into Training and Testing set

```python

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
```

```python
lr = LinearRegression()
lr.fit(x_train,y_train)

rf = RandomForestRegressor()
rf.fit(x_train,y_train)

gb = GradientBoostingRegressor()
gb.fit(x_train,y_train)

xgboost = XGBRegressor()
xgboost.fit(x_train,y_train)
```

Prediction of test dataset

```python
y_p1 = lr.predict(x_test)
y_p2 = rf.predict(x_test)
y_p3 = gb.predict(x_test)
y_p4 = xgboost.predict(x_test)
```

Analysis of the algorithms

```python
s1 = metrics.r2_score(y_test,y_p1)
s2 = metrics.r2_score(y_test,y_p2)
s3 = metrics.r2_score(y_test,y_p3)
s4 = metrics.r2_score(y_test,y_p4)
```

```python
print(s1,s2,s3,s4)
```

```python
R2_Score = ['Linear Regression','Random Forest','Gradient Boosting','XGBoost']
final_Value = pd.DataFrame({'Models':R2_Score,'R2 Score':[s1, s2, s3, s4]})
```

```python
final_Value
```

Plotting the values of Models along X-Axis and R2 Score along Y-Axis

```python
sns.barplot(x='Models',y='R2 Score',data=final_Value)
```
